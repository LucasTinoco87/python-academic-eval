## 🔹 Questão 1 — Classificação de Temperaturas

Dada a lista de temperaturas registradas durante a semana:
temperaturas = [18, 22, 30, 35, 28, 15, 40]
Classifique cada dia como:
•	Abaixo de 20 → "Frio"
•	Entre 20 e 30 → "Agradável"
•	Acima de 30 → "Quente"

Tarefas:

1.	Criar uma lista com as classificações

2.	Contar quantos dias houve em cada categoria


```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]

classificacoes = []
frio_count = 0
agradavel_count = 0
quente_count = 0

for temp in temperaturas:
    if temp < 20:
        classificacao = "Frio"
        frio_count += 1
    elif 20 <= temp <= 30:
        classificacao = "Agradável"
        agradavel_count += 1
    else: # temp > 30
        classificacao = "Quente"
        quente_count += 1
    classificacoes.append(classificacao)

print(f"Classificações: {classificacoes}")
print(f"Número de dias 'Frio': {frio_count}")
print(f"Número de dias 'Agradável': {agradavel_count}")
print(f"Número de dias 'Quente': {quente_count}")
```

## 🔷 Questão 2 — Sistema de Avaliação de Alunos

Dada a lista de notas finais:
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
Classifique:
•	< 5 → "Reprovado"
•	5 a 7 → "Recuperação"
•	7 → "Aprovado"

Tarefas:

1.	Gerar lista de classificações
2.	Calcular:
o	Quantidade de aprovados
o	Percentual de reprovados


```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

classificacoes_alunos = []
aprovados_count = 0
reprovados_count = 0

for nota in notas:
    if nota < 5:
        classificacao = "Reprovado"
        reprovados_count += 1
    elif 5 <= nota < 7: # As per instruction '5 a 7', assuming 7 is approved
        classificacao = "Recuperação"
    else: # nota >= 7 (As per instruction '7 -> Aprovado', assuming 7 and above)
        classificacao = "Aprovado"
        aprovados_count += 1
    classificacoes_alunos.append(classificacao)

total_alunos = len(notas)
percentual_reprovados = (reprovados_count / total_alunos) * 100 if total_alunos > 0 else 0

print(f"Classificações dos alunos: {classificacoes_alunos}")
print(f"Quantidade de alunos aprovados: {aprovados_count}")
print(f"Percentual de alunos reprovados: {percentual_reprovados:.2f}%")
```

## 🔷 Questão 3 — Monitoramento de Consumo de Energia

Consumo diário (kWh):
consumo = [120, 95, 200, 180, 75, 220, 160]
Regras:
•	< 100 → "Baixo"
•	100–180 → "Moderado"
•	180 → "Alto"



Tarefas:

1.	Filtrar dias com consumo alto
2.	Calcular:
o	Total de consumo da semana e
*a*	Média semanal
3.	Exibir alerta se houver mais de 2 dias com consumo alto


```python
consumo = [120, 95, 200, 180, 75, 220, 160]

# 1. Filtrar dias com consumo alto
dias_consumo_alto = [c for c in consumo if c >= 180]

# 2. Calcular total de consumo da semana e média semanal
total_consumo_semanal = sum(consumo)
media_consumo_semanal = total_consumo_semanal / len(consumo)

# 3. Exibir alerta se houver mais de 2 dias com consumo alto
alerta_consumo_alto = False
if len(dias_consumo_alto) > 2:
    alerta_consumo_alto = True

print(f"Dias com consumo alto (>= 180 kWh): {dias_consumo_alto}")
print(f"Total de consumo da semana: {total_consumo_semanal} kWh")
print(f"Média semanal de consumo: {media_consumo_semanal:.2f} kWh")

if alerta_consumo_alto:
    print("ALERTA: Mais de 2 dias com consumo alto esta semana!")
else:
    print("Nenhum alerta de consumo alto excessivo.")
```

## 🔷 Questão 4 — Simulação de Carrinho de Compras

Lista de preços:
precos = [50, 120, 30, 200, 80, 15]
Regras de desconto:
•	Item < 50 → 5% desconto
•	50–150 → 10% desconto
•	150 → 15% desconto

Tarefas:

1.	Criar lista com preços finais
2.	Calcular:
o	Total da compra com desconto
e a Economia total obtida


```python
precos = [50, 120, 30, 200, 80, 15]

precos_finais = []
total_desconto = 0

for preco in precos:
    desconto_aplicado = 0
    if preco < 50:
        desconto_aplicado = preco * 0.05  # 5% desconto
    elif 50 <= preco < 150:
        desconto_aplicado = preco * 0.10  # 10% desconto
    else: # preco >= 150
        desconto_aplicado = preco * 0.15  # 15% desconto

    preco_final = preco - desconto_aplicado
    precos_finais.append(f"{preco_final:.2f}")
    total_desconto += desconto_aplicado

total_compra_com_desconto = sum([float(p) for p in precos_finais])
economia_total_obtida = total_desconto

print(f"Preços originais: {precos}")
print(f"Preços finais com desconto: {precos_finais}")
print(f"Total da compra com desconto: {total_compra_com_desconto:.2f}")
print(f"Economia total obtida: {economia_total_obtida:.2f}")
```
